**Do we need to learn both?**

````python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Define the sizes of each subset
venn_labels = {
    '100': 1,  # Builder Side Data Scientist only
    '010': 1,  # User Side Software Developers only
    '110': 1,  # Builder Side Data Scientist & AI Engineer
    '101': 0,  # Builder Side Data Scientist & User Side Software Developers
    '011': 0,  # User Side Software Developers & AI Engineer
    '111': 1   # All three: Builder Side Data Scientist, AI Engineer, User Side Software Developers
}

# Create the plot
plt.figure(figsize=(8,8))
venn_diagram = venn3(subsets=venn_labels, set_labels=())

# Set custom labels inside the circles
venn_diagram.get_label_by_id('100').set_text('Builder Side\nData Scientist')
venn_diagram.get_label_by_id('010').set_text('User Side\nSoftware Developers')
venn_diagram.get_label_by_id('110').set_text('AI Engineer')
# venn_diagram.get_label_by_id('111').set_text('AI Engineer')  # Overlap of all three
# other overlaps can be labeled if needed

plt.title("Do we need to learn both?")
plt.show()
````
